```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew
import seaborn as sns
```

```python
#generating data
data1 = np.random.normal(loc=0, scale=1, size=1000)  # Normally distributed data
skewness = skew(data1)
print(f'Skewness: {skewness}')
```

```python
sns.histplot(data=data1, kde=True)
```

```python
data2 = pd.DataFrame({
    'values': np.random.lognormal(mean=0, sigma=1, size=1000)  # Right-skewed data
})

skewness = data2['values'].skew()
skewness
```

```python
sns.histplot(data=data2, kde=True)
```

```python
#Right-Skewed Distribution
data = np.random.lognormal(mean=0, sigma=1, size=1000)
```

```python
mean = np.mean(data)
median = np.median(data)
mode = max(set(data), key=list(data).count)
```

```python
plt.figure(figsize=(10, 4))
plt.hist(data, bins=30, alpha=0.6, color='blue', label='Data')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='yellow', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')

plt.title('Right-Skewed Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
skewness = skew(data)
print(f'Skewness: {skewness}')
```

```python
#Left-Skewed Distribution
data = -np.random.lognormal(mean=0, sigma=0.7, size=1000)
```

```python
mean = np.mean(data)
median = np.median(data)
mode = min(set(data), key=list(data).count)
```

```python
plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, alpha=0.6, color='purple', label='Data')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='yellow', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')

plt.title('Left-Skewed Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
skewness = skew(data)
print(f'Skewness: {skewness}')
```

```python
#Normal Distribution where the skewness is approximately zero
data_normal = np.random.normal(loc=0, scale=1, size=1000)

mean_normal = np.mean(data_normal)
median_normal = np.median(data_normal)
mode_normal = min(set(data_normal), key=list(data_normal).count)

plt.figure(figsize=(10, 6))
plt.hist(data_normal, bins=30, alpha=0.6, color='gray', label='Data')
plt.axvline(mean_normal, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_normal:.2f}')
plt.axvline(median_normal, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {median_normal:.2f}')
plt.axvline(mode_normal, color='yellow', linestyle='dashed', linewidth=2, label=f'Mode: {mode_normal:.2f}')

plt.title('Normal Distribution (Skewness ~ 0)')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
skewness = skew(data_normal)
print(f'Skewness: {skewness}')
```
